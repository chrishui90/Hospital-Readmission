# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

# B. Provide support for your arguments and your own recommendations with a statistically sound analysis:

```python
#groupA - <=100 Discharges
clean_hospital_read_df = hospital_read_df[(hospital_read_df['Number of Discharges'] != 'Not Available')]
filt = clean_hospital_read_df['Number of Discharges'].astype(int) <=100
groupA = clean_hospital_read_df[filt]
filt2 = groupA['Number of Discharges'].astype(int) != 0
groupA = groupA[filt2]
groupA.head()
```

```python
#groupB - >1000 Discharges
clean_hospital_read_df = hospital_read_df[(hospital_read_df['Number of Discharges'] != 'Not Available')]
filt = clean_hospital_read_df['Number of Discharges'].astype(int) >=1000
groupB = clean_hospital_read_df[filt]
filt2 = groupB['Number of Discharges'].astype(int) != 0
groupB = groupB[filt2]
groupB.head()
```

```python
print (len(groupA))
print (len(groupB))
```

### Statistical Significance Test (Alpha > 0.01)

We have opted to use a student T-Test.
We are measuring whether the mean(s) vary significantly across our sample populations.

```python
import scipy.stats as stats
result = stats.ttest_ind(groupA['Excess Readmission Ratio'],groupB['Excess Readmission Ratio'], equal_var=False)
print("The P value is :", result[1])
```

With a P value significantly less than 0.01, we can reject the Null Hypothesis that No difference exists in the mean Excess Readmission Ratio in hospitals with discharges <=100 and hospitals with discharges > 1000 

# Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?


Statistical Significance and Practical Significance are two very different beasts.
As Long Beach Calfornia states it:
Statistical significance shows the mathematical probability that a relationship between two or more variables exists, while practical significance refers to relationships between variables with real-world applications.

Often within Engineering, our variables which are statistically significant are often grounded in practicality.
i.e. If we want to know the key factor(s) that are contribution to pipe corrosion and the associated weightings, this can be done via a combination of flow rate, maintenance regime, pipe material etc. 

However, even if pipe material was to be picked up as statistically significant, it isn't likely we will end up changing the pipes constituion (we won't), so the relationship is statistically significant, but practically, we wouldn't do anything about it. 

In the context of our example, we have shown that there exists statistically significant differences in mean reeadmission rates between large and small hospitals. In this case, whether it means anything to the client however is dependent on client SME knowledge. This observation is readily observible even without the need for a statistical test.
